```python
import json
import csv
JSON_Dosyası='envanter.json'
def envanteri_yükle():
  try:
    with open(JSON_Dosyası,'r',encoding='utf-8') as dosya:
      return json.load(dosya)
  except FileNotFoundError:
    return {}

def envanteri_kaydet(envanter):
  with open(JSON_Dosyası,'w',encoding='utf-8') as dosya:
    json.dump(envanter,dosya,ensure_ascii=False,indent=4)

def urun_ekle(envanter):
  urun_adi=input("Eklenecek ürünün adı:").strip().capitalize()
  if urun_adi in envanter:
    print(f"Hata:'{urun_adi}' zaten envanterde mevcut!")
    return

  try:
    adet=int(input(f"{urun_adi} için stok adeti:"))
    fiyat=float(input(f"{urun_adi} için birim fiyatı:"))
    envanter[urun_adi]={"Adet":adet,"fiyat":fiyat}
    envanteri_kaydet(envanter)
    print(f"Başarılı:'{urun_adi}' envantere eklendi ve kaydedildi.")
  except ValueError:
    print("Hata:Geçersiz veri girişi!")

def urun_sil(envanter):
  urun_adi=input("Silinecek ürünün adı:").strip().capitalize()
  if urun_adi not in envanter:
    print(f"Hata:'{urun_adi}' envanterde yok!")
    return

  del envanter[urun_adi]
  envanteri_kaydet(envanter)

def envanteri_görüntüle(envanter):
  print("-----MEVCUT ENVANTER-----")
  if not envanter:
    print("Hata:Envanter boş!")
    return
  else:
    for urun_adi, bilgiler in envanter.items():
      print(f"{urun_adi}: Adet={bilgiler['Adet']}, Fiyat={bilgiler['fiyat']:.2f} TL")
  print("--------------------------\n")

def urun_güncelle(envanter):
  urun_adi=input("Güncellenecek ürünün adı:").strip().capitalize()
  if urun_adi not in envanter:
    print(f"Hata:'{urun_adi}' envanterde yok!")
    return
  try:
    yeni_adet=int(input(f"{urun_adi} için yeni stok adeti:"))
    yeni_fiyat=float(input(f"{urun_adi} için yeni birim fiyatı:"))
    envanter[urun_adi]["Adet"] = yeni_adet
    envanter[urun_adi]["fiyat"] = yeni_fiyat
    envanteri_kaydet(envanter)
    print(f"Başarılı:'{urun_adi}' envanterde güncellendi ve kaydedildi.")
  except ValueError:
    print("Hata:Geçersiz veri girişi!")

def envanter_raporu(envanter):
  print("\n------ENVANTER RAPORU------")
  if not  envanter:
    print("Hata:Envanter boş!")
    return
  else:
    toplam_fiyat=0.0
    toplam_urun_cesidi=len(envanter)
    toplam_urun_adedi=0
    for bilgiler in envanter.values():
      toplam_urun_adedi+=bilgiler['Adet']
      toplam_fiyat+=bilgiler['fiyat']*bilgiler['Adet']

    print(f"Toplam ürün çeşidi:{toplam_urun_cesidi}")
    print(f"Toplam ürün adedi:{toplam_urun_adedi}")
    print(f"Toplam fiyat:{toplam_fiyat:.2f} TL")
    print("------------------------------\n")

def envanteri_csv_dosyasına_aktar(envanter):
  if not envanter:
    print("Hata:Envanter boş!")
    return
  try:
    with open('envanter.csv','w',newline='',encoding='utf-8') as dosya:
      yazici=csv.writer(dosya)
      yazici.writerow(['Ürün Adı','Adet','Fiyat'])
      for urun_adi,bilgiler in envanter.items():
        yazici.writerow([urun_adi,bilgiler['Adet'],bilgiler['fiyat']])
    print("BAŞARILI: envanter 'envanter.csv' dosyasına aktarıldı.")

  except Exception as hata:
    print(f"Hata:CSV dosyasına yazma sırasında bir hata oluştu:{hata}")

def main():
  envanter=envanteri_yükle()

  while True:
    print("\n-----MENÜ-----")
    print("1-Ürün Ekle")
    print("2-Ürün Sil")
    print("3-Envanteri Görüntüle")
    print("4-Ürün Güncelle")
    print("5-Envanter Raporu")
    print("6-Envanteri CSV Dosyasına Aktarı")
    print("7-Çıkış")

    secim=input("Seçiminiz:").strip()
    if secim=='1':
      urun_ekle(envanter)
    elif secim=='2':
      urun_sil(envanter)
    elif secim=='3':
      envanteri_görüntüle(envanter)
    elif secim=='4':
        urun_güncelle(envanter)
    elif secim=='5':
        envanter_raporu(envanter)
    elif secim=='6':
      envanteri_csv_dosyasına_aktar(envanter)
    elif secim=='7':
      print("Çıkış yapılıyor...")
      break
    else:
      print("Hata:Geçersiz seçim!")

if __name__=='__main__':
  main()
```
