```python
#mini günlük
from datetime import datetime
def gunluk_yaz():
  with open("gunluk.txt","a",encoding="utf-8")as dosya:
    yaz=input("Bugün günlüğe ne yazmak istersin?")
    dosya.write(datetime.now().strftime("%Y-%m-%d %H:%M")+"\n" + yaz+"\n") # Added timestamp and corrected write
  print("günlüğe eklendi.")

def gunluk_oku():
  try: # Added try-except for FileNotFoundError
    with open("gunluk.txt","r",encoding="utf-8")as dosya:
      oku=dosya.read()
      print(oku,"\n")
  except FileNotFoundError:
    print("Günlük dosyası bulunamadı.") # Informative message

def gunluk_ara():
  kelime=input("Aranacak kelime:")
  bulundu = False # Kelimenin bulunup bulunmadığını takip etmek için bayrak
  try: # Added try-except for FileNotFoundError
    with open("gunluk.txt","r",encoding="utf-8")as dosya:
      for satir in dosya:
        if kelime in satir:
          print(satir.strip()) # Satır sonundaki boşluğu temizlemek için strip() kullanıldı
          bulundu = True # Kelime bulundu, bayrağı True yap
  except FileNotFoundError:
    print("Günlük dosyası bulunamadı.") # Informative message

  if not bulundu: # Döngü bittikten sonra kelime bulunamadıysa
    print("kelime bulunamadı.")

def gunluk_temizle():
  onay=input("günlüğü temizlemek istiyor musunuz? (e/h)")
  if onay.lower()=="e": # Used .lower() for case-insensitive check
    try: # Added try-except for potential issues
      with open("gunluk.txt","w",encoding="utf-8")as dosya:
        dosya.write("")
        print("dosya temizlendi.")
    except Exception as e: # Catch any potential error during file operations
      print(f"Bir hata oluştu: {e}")
  else:
    print("işlem iptal edildi.")


def menu():
  while True:
    print("--------Günlük--------")
    print("1-Günlük yaz")
    print("2-Günlük oku")
    print("3-Çıkış")
    print("4-Günlük ara") # Added search option to menu
    print("5-Günlük temizle") # Added clean option to menu
    secim=input("Seçiminiz:")

    if secim=="1":
      gunluk_yaz()
    elif secim=="2":
      gunluk_oku()
    elif secim=="3":
      print("çıkış yapıldı.")
      break
    elif secim=="4":
      gunluk_ara() # Call the updated gunluk_ara function
    elif secim=="5":
      gunluk_temizle() # Call the gunluk_temizle function
    else:
      print("Yanlış seçim yaptınız.")

menu()
```
