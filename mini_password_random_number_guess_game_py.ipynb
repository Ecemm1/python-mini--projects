```python
#mini şifre ile girilen tahmin oyunu

 # ilk olarak mini şifre uygulaması yapıp ardından mini bir oyun yapacağım ve sonra mini şifreli tahmin oyunu yazalım.

kullanicilar={"ecem":"12345", "ezgi":"1234","yusuf":"123"}
hak=5


while hak>0:
  kullanici=input("kullanıcı adınızı giriniz:")
  sifre=input("şifrenizi giriniz:")

  if kullanici in kullanicilar and sifre==kullanicilar[kullanici]:
    print("hoşgeldin",kullanici)
    break
  else:
    print("kullanıcı adı veya şifre yanlış...")

  hak-=1
  print(hak,"hakkın kaldı")
if hak==0:
  print("hakkın bitti!")

#sayı tahmin oyunu

import random
sayi=random.randint(1,100)
hak=5
print("Sayı tahmin oyununa hoşgeldiniz...")
print("hadi, 1 ile 100 arası tutulan sayıyı tahmin edelim :)")
while hak>0:
  tahmin=int(input("bir sayı tahmin et:"))
  if tahmin==sayi:
    print("teşekkürler, doğru bildin!")
    break
  elif tahmin<sayi:
    print("daha büyük bir sayı tahmin et")
  else:
    print("daha küçük bir sayı tahmin et")
  hak-=1
  print(hak,"hakkın kaldı")
  if hak==0:
    print("hakkın bitti!")
    print("tutulan sayi:",sayi)


#iki mini projeyi birleştirelim
import random

# Create skorlar.txt if it doesn't exist
DOSYA="skorlar.txt"
try:
    with open("skorlar.txt", "x") as f:
        pass
except FileExistsError:
    pass


kullanicilar={}
adet=int(input("kullanıcı sayısını giriniz:"))

for i in range(adet):
  isim=input(f"{i+1}.kullanıcının adını gir:")
  sifre=input(f"{i+1}.kullanıcının şifresini gir:")
  kullanicilar[isim]=sifre

print("\n kullanıcı listesi oluşturuldu.")
#skorlar için sözlük
skorlar={}

# Load existing scores from the file
try:
    with open("skorlar.txt", "r") as f:
        for line in f:
            isim, puan = line.strip().split(":")
            skorlar[isim] = int(puan)
except FileNotFoundError:
    pass # File doesn't exist yet, continue with empty skorlar


#giriş ekranı
print("\n-------giriş ekranı--------")
giris_basarili = False
aktif_kullanici=""
hak = 3 # Giriş denemesi için hak sayısı

while hak > 0:
    isim = input("kullanıcı adınızı giriniz:")
    sifre = input("şifrenizi giriniz:")

    if isim in kullanicilar and sifre == kullanicilar[isim]:
        print("hoşgeldin", isim)
        aktif_kullanici=isim
        giris_basarili = True
        break
    else:
        print("kullanıcı adı veya şifre yanlış...")
        hak -= 1
        print(hak, "hakkın kaldı")

if giris_basarili:
    while True: # Outer loop for playing multiple games
        # oyun ekranı
        print("\n-------oyun ekranı-------")
        sayi = random.randint(1, 100)
        hak = 5 # Tahmin oyunu için hak sayısı
        puan=0
        print("Sayı tahmin oyununa hoşgeldiniz...")
        print("hadi, 1 ile 100 arası tutulan sayıyı tahmin edelim :)")

        while hak > 0:
            try:
                tahmin = int(input("bir sayı tahmin et:"))
                if tahmin == sayi:
                    print("tebrikler, doğru bildin!")
                    puan = hak * 10
                    print("puanın:",puan)
                    break
                elif tahmin < sayi:
                    print("daha büyük bir sayı tahmin et")
                else:
                    print("daha küçük bir sayı tahmin et")
            except ValueError:
                print("Geçerli bir sayı giriniz.")
                continue # Geçersiz girişte hakkı düşürme

            hak -= 1
            print(hak, "hakkın kaldı")

        if hak == 0:
            print("hakkın bitti!")
            print("tutulan sayi:", sayi)

        #skoru kaydet
        if aktif_kullanici  in skorlar:
         skorlar[aktif_kullanici]+=puan

        else :
         skorlar[aktif_kullanici]=puan


        print("skorlar:",skorlar)

        # Write updated scores back to the file
        with open("skorlar.txt", "w") as f:
            for isim, puan in skorlar.items():
                f.write(f"{isim}:{puan}\n")

        print("\n -------LİDER TABLOSU--------")
        # Read and display scores from the file
        try:
            with open("skorlar.txt", "r") as f:
                skor_listesi = []
                for satir in f:
                    isim, puan = satir.strip().split(":")
                    skor_listesi.append((isim, int(puan)))
                skor_listesi.sort(key=lambda x: x[1], reverse=True)
                for sira, (isim, puan) in enumerate(skor_listesi, start=1):
                    print(f"{sira}.{isim}:{puan}")
        except FileNotFoundError:
             print("Skor dosyası bulunamadı.")


        tekrar = input("tekrar oynamak ister misin?(e/h)").lower()
        if tekrar != "e":
            print("güle güle")
            break # Exit the outer game loop if the user doesn't want to play again


else:
    print("Giriş hakkınız bitti!")
```
